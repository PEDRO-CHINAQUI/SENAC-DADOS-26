```python
dados = np.array([12, 15, 17, 20, 22, 25, 28, 30, 35, 40])
q1 = np.percentile(dados, 25)
q2 = np.percentile(dados, 50)
q3 = np.percentile(dados, 75)
print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2, Mediana): {q2}')
print(f'Terceiroquartil (Q3): {q3}')
```

```python
array_furto_celular = np.array(df_furto_celular_cisp['furto_celular'])
q1_furto_celular = np.percentile(array_furto_celular, 25)
q2_furto_celular = np.percentile(array_furto_celular, 50)
q3_furto_celular = np.percentile(array_furto_celular, 75)
media = np.mean(array_furto_celular)
mediana = np.median(array_furto_celular)
distancia = (media - mediana) / mediana
iqr = q3_furto_celular - q1_furto_celular
limite_superior = q3_furto_celular + (1.5 * iqr)
limite_inferior = q1_furto_celular - (1.5 * iqr)
print(f'Primeiro quartil (Q1): {q1_furto_celular}')
print(f'Segundo quartil (Q2, Mediana): {q2_furto_celular}')
print(f'Terceiroquartil (Q3): {q3_furto_celular}')
print(f'O valor da média é: {media}')
print(f'O valor da mediana é: {mediana}')
print(f'A distância entre a média e a mediana é: {distancia * 100}')
print(f'O limite superior para detecção de outliers é: {limite_superior}')
print(f'O limite inferior para detecção de outliers é: {limite_inferior}')
```

```python
 
outliers_cisp_furto_celular = df_furto_celular_cisp.loc[df_furto_celular_cisp['furto_celular'] >= limite_superior]

outliers_cisp_furto_celular = outliers_cisp_furto_celular.sort_values (by='furto_celular', ascending=False)

outliers_cisp_furto_celular

```

```python
df_capital = df[df['regiao'] == 'Capital'] 
df_veiculos_cisp = df_capital.groupby('cisp')[['furto_veiculos', 'roubo_veiculo']].sum().reset_index()
df_veiculos_cisp['total_veiculos'] = df_veiculos_cisp['furto_veiculos'] + df_veiculos_cisp['roubo_veiculo']
array_veiculos = np.array(df_veiculos_cisp['total_veiculos'])
q1_veiculos = np.percentile(array_veiculos, 25)
q3_veiculos = np.percentile(array_veiculos, 75)
iqr_veiculos = q3_veiculos - q1_veiculos
limite_superior_veiculos = q3_veiculos + (1.5 * iqr_veiculos)
outliers_veiculos_cisp = df_veiculos_cisp[df_veiculos_cisp['total_veiculos'] > limite_superior_veiculos].sort_values(by='total_veiculos', ascending=False)
outliers_veiculos_cisp
```
