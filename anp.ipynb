```python
import pandas as pd
import numpy as np
```

```python
dados_1sem = 'Preços semestrais - AUTOMOTIVOS_2025.01.csv'
df_1sem = pd.read_csv(dados_1sem, encoding='latin1', sep=';')
df_1sem.head(3)
```

```python
dados_2sem = 'Preços semestrais - AUTOMOTIVOS_2025.02.csv'
df_2sem = pd.read_csv(dados_2sem, sep=';')
df_2sem.head(3)
```

```python
df_anp = pd.concat([df_1sem, df_2sem], ignore_index= True)
```

```python
df_anp['Valor de Venda'] = df_anp['Valor de Venda'].str.replace(',','.').astype(float)
```

```python
df_gasolina = df_anp.loc[df_anp['Produto'] == 'GASOLINA']
df_etanol = df_anp.loc[df_anp['Produto'] == 'ETANOL']
```

```python
df_media_est_gasolina= df_gasolina.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index()
df_media_est_gasolina = df_media_est_gasolina.sort_values(by='Valor de Venda', ascending=False)

df_media_est_gasolina 
```

```python
df_media_est_etanol = df_etanol.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index()
df_media_est_etanol = df_media_est_etanol.sort_values(by='Valor de Venda', ascending=False)

df_media_est_etanol 
```

```python
top5_gasolina = df_media_est_gasolina.head(5)
print("Top 5 Estados com Maiores Preços Médios de Gasolina:")
print(top5_gasolina)

top5_etanol = df_media_est_etanol.head(5)
print("\nTop 5 Estados com Maiores Preços Médios de Etanol:")
print(top5_etanol)
```

```python
df_media_reg_gasolina = df_gasolina.groupby('Regiao - Sigla')['Valor de Venda'].mean().reset_index()
df_media_reg_gasolina = df_media_reg_gasolina.sort_values(by='Valor de Venda', ascending=False)

df_media_reg_etanol = df_etanol.groupby('Regiao - Sigla')['Valor de Venda'].mean().reset_index()
df_media_reg_etanol = df_media_reg_etanol.sort_values(by='Valor de Venda', ascending=False)

print("Médias de Preços por Região - Gasolina:")
print(df_media_reg_gasolina)

print("\nMédias de Preços por Região - Etanol:")
print(df_media_reg_etanol)
```
